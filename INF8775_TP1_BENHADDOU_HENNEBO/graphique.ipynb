```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
# All datas
N = [2, 3, 4, 5, 6, 7, 8, 9, 10]
sizes = [pow(2,x) for x in N]
data = np.loadtxt("resultats.txt").astype(int)

# 2^5 to 2^10
N = N[3:]
sizes = sizes[3:]
datas = []
for r in data:
    datas.append(r[3:])
```

```python
fig, ax = plt.subplots(1, 3, figsize = (17,5))

ax[0].plot(sizes, datas[0], label="Conventionnel",linewidth = 2, color = "c", marker='o', markersize = 7)
ax[1].plot(sizes, datas[1], label="Strassen",linewidth = 2, color = "b", marker='o', markersize = 7)
ax[2].plot(sizes, datas[2], label="Strassen - Seuil",linewidth = 2, color = "g", marker='o', markersize = 7)

for i in range(3):
    ax[i].set_xlabel("Taille (log)")
    ax[i].set_ylabel("Temps (log)")
    ax[i].set_xscale("log", base=2)
    ax[i].set_yscale("log", base=2)

    ax[i].set_title("Test de puissance")
    ax[i].grid()
    ax[i].legend()

plt.show()

print("m = 3.53808")
print("m = 2.78515 ")
print("m = 2.76974")
```

```python
m0 = 3.53808
m1 = 2.78515
m2 = 2.76974
hx0 = [pow(size,m0) for size in sizes]
hx1 = [pow(size,m1) for size in sizes]
hx2 = [pow(size,m2) for size in sizes]
y0 = [0] * len(hx0)
y1 = [0] * len(hx1)
y2 = [0] * len(hx2)

for i,x in enumerate(hx0):
    y0[i] = datas[0][i]/hx0[i]
    y1[i] = datas[1][i]/hx1[i]
    y2[i] = datas[2][i]/hx2[i]

fig, ax = plt.subplots(1, 3, figsize = (17,5))

ax[0].plot(sizes, y0, label="Conventionnel",linewidth = 2, color = "c", marker='o', markersize = 7)
ax[1].plot(sizes, y1, label="Strassen",linewidth = 2, color = "b", marker='o', markersize = 7)
ax[2].plot(sizes, y2, label="Strassen - Seuil",linewidth = 2, color = "g", marker='o', markersize = 7)

for i in range(3):
    ax[i].set_xlabel("Taille")
    ax[i].set_ylabel("Temps")

    ax[i].set_title("Test du rapport")
    ax[i].grid()
    ax[i].legend()

plt.show()

print("a = 1.495565937725573*10^-7")
print("a = 0.000461096158896 ")
print("a = 0.000012019703461574022")
```
