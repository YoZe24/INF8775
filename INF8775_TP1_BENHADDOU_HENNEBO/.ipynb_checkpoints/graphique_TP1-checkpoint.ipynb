```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
# All datas
N = [2, 3, 4, 5, 6, 7, 8, 9, 10]
sizes = [pow(2,x) for x in N]
data = np.loadtxt("resultats.txt").astype(int)

# 2^5 to 2^10
N_2 = N[3:]
sizes_2 = sizes[3:]
data_2 = []
for r in data:
    data_2.append(r[3:])
```

```python
fig, ax = plt.subplots()

ax.plot(sizes, data[0], label="Conventionnel",linewidth = 2, color = "c", marker='o', markersize = 7)
ax.plot(sizes, data[1], label="Strassen",linewidth = 2, color = "b", marker='o', markersize = 7)
ax.plot(sizes, data[2], label="Strassen - Seuil",linewidth = 2, color = "g", marker='o', markersize = 7)

ax.set_xlabel("Taille")
ax.set_ylabel("Temps")
ax.set_xscale("log", base=2)
ax.set_yscale("log", base=2)

ax.grid()
ax.legend()

# for X, Y in zip(sizes, data[0]):
#     ax.annotate('{}'.format(round(Y,2)), xy=(X,Y), xytext=(-5, -15), ha='left',textcoords='offset points')
# for X, Y in zip(sizes, data[1]):
#     ax.annotate('{}'.format(round(Y,2)), xy=(X,Y), xytext=(5, 8), ha='left',textcoords='offset points')
# for X, Y in zip(sizes, data[2]):
#     ax.annotate('{}'.format(round(Y,2)), xy=(X,Y), xytext=(5, 8), ha='left',textcoords='offset points')

plt.show()
```
